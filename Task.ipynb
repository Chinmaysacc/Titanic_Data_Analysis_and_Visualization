```python
import pandas as pd
df = pd.read_csv(r"D:\ELEVATE LABS\MINI TASK 5\train.csv")
```

```python
# Basic Structure
print("Basic Structure")
print("")
print(df.shape)  # rows and columns
print(df.columns)
print(df.dtypes)
```

```python
# Missing values
print("Missing Values")
print("")
print(df.isnull().sum())
```

```python
# Summary statistics
print("Summary statistics")
print("")
print(df.describe(include='all'))
```

```python
# Value counts for categorical variables
print("Value counts for categorical variables")
for col in df.select_dtypes(include='object').columns:
    print(f"Value counts for {col}:\n{df[col].value_counts()}\n")
```

```python
# Correlation matrix (numeric only)
print("Correlation matrix (numeric only)")
print("")
correlation = df.select_dtypes(include=['number']).corr()
print("Correlation matrix:\n", correlation)
```

```python
#Visualize Using Matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

plt.bar(df['Survived'].value_counts().index, df['Survived'].value_counts().values, color=['red', 'green'])
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.title("Survival Count")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
gender_counts = df['Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightpink'])
plt.title("Gender Distribution")
plt.axis('equal')
plt.show()
```

```python
plt.hist(df['Age'].dropna(), bins=30, color='orange', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
pclass_survived = pd.crosstab(df['Pclass'], df['Survived'])
pclass_survived.plot(kind='bar', stacked=True, color=['darkred', 'green'])
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python
plt.scatter(df['Age'], df['Fare'], alpha=0.6, c=df['Survived'], cmap='coolwarm')
plt.title("Age vs Fare Colored by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.colorbar(label='Survived (0 = No, 1 = Yes)')
plt.show()
```

```python
age_survived_0 = df[df['Survived'] == 0]['Age']
age_survived_1 = df[df['Survived'] == 1]['Age']

plt.boxplot([age_survived_0.dropna(), age_survived_1.dropna()], labels=['Not Survived', 'Survived'])
plt.title("Age Distribution by Survival")
plt.ylabel("Age")
plt.show()
```

```python
embarked_counts = df['Embarked'].value_counts()
plt.barh(embarked_counts.index, embarked_counts.values, color='goldenrod')
plt.title("Passengers by Embarkation Port")
plt.xlabel("Number of Passengers")
plt.ylabel("Port")
plt.show()
```

```python
#Visualize using Seaborn
```

```python
import seaborn as sns
```

```python
sns.countplot(data=df, x='Survived', palette='Set2')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.title("Survival Count")
plt.show()
```

```python
sns.histplot(data=df, x='Age', bins=30, kde=True, color='orange')
plt.title("Age Distribution with KDE")
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived', palette='coolwarm')
plt.title("Survival by Passenger Class")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
sns.boxplot(data=df, x='Survived', y='Age', palette='husl')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.violinplot(data=df, x='Pclass', y='Fare', hue='Survived', split=True, palette='muted')
plt.title("Fare Distribution by Class and Survival")
plt.show()
```

```python
sns.countplot(data=df, x='Embarked', hue='Sex', palette='cubehelix')
plt.title("Embarkation Port by Gender")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
